## **Problem 1: Plotting the Density of Earth**
In this folder, notice that there is a text file labeled 'PREM_EarthModel.txt'. This provides a 1-D model representing the average properties of the Earth as a function of radius/depth. Your task is to read in the text file using `astropy.table` (code provided) and create a plot of radius from the center of earth (x-axis) versus density of the earth (y-axis). Be sure to add a title, axes labels, and a legend!

```python
# imports (just run this cell)
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
# reading in the text file (just run this cell)
filepath = 'PREM_EarthModel.txt'
table = Table.read(filepath, format='ascii')
table
```

```python
# your plotting code goes here!
radii = table["radius"]
densities = table["density"]
plt.figure()
plt.plot(radii, densities, label = "density")
plt.title("Density vs. Radius of Earth")
plt.xlabel("Radius (km)")
plt.ylabel(r"$Density (g/cm^3)$")
plt.grid(True)
plt.legend()
plt.show()
```
